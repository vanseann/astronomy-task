```python
# Import Library yang dibutuhkan
import random as rd
import numpy as np

def monte_carlo(interval):
    # Definisikan variabel yang akan digunakan
    n = 0 # n adalah jumlah titik yang dihasilkan di dalam lingkaran
    m = 0 # m adalah jumlah titik yang berada dalam persegi
    
    # Jumlah angka acak yang dihasilkan = probabilitas x
    for i in range(interval):
        x = rd.uniform(-1, 1) # Nilai x dan y dihasilkan secara acak dari distribusi uniform
        y = rd.uniform(-1, 1) # Rentang nilai x dan y adalah -1 hingga 1
        if x**2 + y**2 <= 1: # Gunakan fungsi if untuk memeriksa apakah (x, y) terletak di dalam lingkaran
            n += 1
        m += 1
        
        # Memperkirakan nilai pi,
        pi_mc = 4* n/m # pi_mc = 4*(jumlah titik yang dihasilkan di dalam lingkaran)/(jumlah titik yang berada dalam
    return pi_mc

# Membuat fungsi untuk mencari galat

def error(pi_mc):
    pi_asli = np.pi
    galat = abs(pi_asli - pi_mc)
    return galat

# Membuat list kosong yang nanti akan diisi dengan galat-galat dari interval yang berbeda
g = []

# Definisikan interval sebagai berapa kali iterasi
intervals = [1000,10000,100000,1000000,10000000]
for interval in intervals:
    result = monte_carlo(interval)
    err = error(result)
    g.append(err)
    print('Nilai pi dengan interval {} adalah : '. format(interval), result)
    print('Nilai galat dengan interval {} adalah : '. format(interval), err)
```

```python
# Membuat grafik galat dari estimasi nilai pi terhadap interval dalam logaritmik
import matplotlib.pyplot as plt

plt.scatter(np.log10(intervals), g, c='slateblue')
plt.title('Plot Interval dalam Logaritmik terhadap Galat pi')
plt.xlabel('Interval dalam Logaritmik')
plt.ylabel('Galat pi')
plt.show()
```
