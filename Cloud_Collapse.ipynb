# **Soal 1**



```python
#library
import numpy as np
import matplotlib.pyplot as plt

#mendefinisikan masing-masing variabel yang diperlukan
rho1 = 5*10**-13
rho2 = 10**-13
rho3 = 5*10**-14
G = 6.67*10**-11

#mendefinisikan X untuk setiap rho
X1 = ((8*np.pi/3)*G*rho1)**(1/2)
X2 = ((8*np.pi/3)*G*rho2)**(1/2)
X3 = ((8*np.pi/3)*G*rho3)**(1/2)


#mendefinisikan theta = r/r0
theta = []
for i in range(1, 10000):
  theta += [i*(1/10000)]

#mendefinisikan nilai t
t1 = []
t2 = []
t3 = []
theta1 = 0
t = 0

for i in range(1, 10000):
  theta1 = theta[i-1]
  Xi = np.arccos(theta1**(1/2))
  t = ((Xi + (1/2)*np.sin(2*Xi))/X1)/(365.25*24*3600)
  t1 += [t]
for i in range(1, 10000):
  theta1 = theta[i-1]
  Xi = np.arccos(theta1**(1/2))
  t = ((Xi + (1/2)*np.sin(2*Xi))/X2)/(365.25*24*3600)
  t2 += [t]
for i in range(1, 10000):
  theta1 = theta[i-1]
  Xi = np.arccos(theta1**(1/2))
  t = ((Xi + (1/2)*np.sin(2*Xi))/X3)/(365.25*24*3600)
  t3 += [t]

#membuat plot r/r0 terhadap waktu
plt.plot(t1, theta, label = 'rho1 = 5 x 10^(-13) kg/m^3', c = 'green')
plt.plot(t2, theta, label = 'rho2 = 10^(-13) kg/m^3', c = 'red')
plt.plot(t3, theta, label = 'rho3 = 5 x 10^(-14) kg/m^3', c = 'blue')
plt.title('Plot Cloud Collapse')
plt.xlabel('time (years)')
plt.ylabel('r/r0')
plt.legend(prop = {'size': 8})
```

# **Soal 2**

```python
#library
import numpy as np
import matplotlib.pyplot as plt

#mendefinisikan masing-masing variabel yang diperlukan
rho1 = 5*10**-13
rho2 = 10**-13
rho3 = 5*10**-14
G = 6.67*10**-11

#mendefinisikan nilai rho/rho_0
rasio_rho = []
log_rasio_rho = []
for i in range (500, 10000):
  rasio = 1/((theta[i-1])**3)   #rasio_rho = rho/rho_0 sama dengan 1/theta^3 karena berbanding terbalik dengan dengan r^3
  log_rasio = np.log(rasio)/np.log(10)
  rasio_rho += [rasio]
  log_rasio_rho += [log_rasio]

#mendefinisikan rasio_t = t/tff
tff1 = ((3*np.pi/(32*G*rho1))**(1/2))/(365.25*24*3600)
tff2 = ((3*np.pi/(32*G*rho2))**(1/2))/(365.25*24*3600)
tff3 = ((3*np.pi/(32*G*rho3))**(1/2))/(365.25*24*3600)

rasio_t1 = []
for i in range(500, 10000):
  rasio = t1[i-1]/tff1
  rasio_t1 += [rasio]
rasio_t2 = []
for i in range(500, 10000):
  rasio = t2[i-1]/tff2
  rasio_t2 += [rasio]
rasio_t3 = []
for i in range(500, 10000):
  rasio = t3[i-1]/tff3
  rasio_t3 += [rasio]

#membuat plot log(rho/rho_0) terhadao t/tff
plt.plot(rasio_t1, log_rasio_rho, label = 'rho1 = 5 x 10^(-13) kg/m^3', c = 'green')
plt.plot(rasio_t2, log_rasio_rho, label = 'rho2 = 10^(-13) kg/m^3', c = 'red')
plt.plot(rasio_t3, log_rasio_rho, label = 'rho3 = 5 x 10^(-14) kg/m^3', c = 'blue')
plt.title('Plot Free-Fall Timescale')
plt.xlabel('t/tff')
plt.ylabel('log(rho/rho_0)')
plt.legend(prop = {'size': 8})
```
