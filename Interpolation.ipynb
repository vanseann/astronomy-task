```python
def interpolation(x,y, xp):
    yp = y[0] + ((y[1]-y[0])/(x[1]-x[0])) * (xp - x[0])
    return yp

x_all = [0, 20, 40, 60, 80, 100]
y_all = [26.0, 48.6, 61.6, 71.2, 74.8, 75.2]

ts=[10,30,50,70,90]
i = 1
j = 0
x = []
y = []
s = 0
ss = []
for t in ts:
  lower = int(t/10) - i
  upper = int(t/10) - j
  x.append(x_all[lower])
  x.append(x_all[upper])
  y.append(y_all[lower])
  y.append(y_all[upper])

  # Cetak hasi interpolasi
  s = interpolation(x,y,t)
  print("Interpolasi t = {} adalah ".format(t), s)
  ss.append(s)
  i = i + 1
  j = j + 1
  x.pop()
  x.pop()
  y.pop()
  y.pop()
```

**Latihan 2**

```python
def interpolate(xp, xs, ys):
  yp = 0
  n = len(xs)

  for i in range(n):

      p = 1

      for j in range(n):
          if i != j:
              p = p * (xp - xs[j])/(xs[i] - xs[j])

      yp = yp + p * ys[i]
  return yp

x_all = [0, 20, 40, 60, 80, 100]
y_all = [26.0, 48.6, 61.6, 71.2, 74.8, 75.2]
xps = [10,30,50,70,90]
yps =[]
for xp in xps:
  yp = interpolate(xp, x_all, y_all)
  print('Interpolasi t = {} adalah '.format(xp), yp)
  yps.append(yp)
```

**Plot perbandingan interpolasi linear dan interpolasi lagrange**

```python
import matplotlib.pyplot as plt

plt.plot(xps, yps, label = 'interpolasi lagrange')
plt.plot(ts, ss, label = 'interpolasi linear')
plt.title('Plot Temperatur Terhadap Waktu')
plt.xlabel('Waktu')
plt.ylabel('Temperatur')
plt.legend()
plt.show()
```

Berdasarkan grafik tersebut, dapat disimpulkan bahwa hasil antara interpolasi linear dan interpolasi langrange memiliki nilai hampir sama. Tetapi, hasil dari interpolasi langrange memiliki tingkat akurasi yang lebih tinggi dibandingkan dengan hasil interpolasi linear.

```python
import numpy as np
import scipy.interpolate
import matplotlib.pyplot as plt
import pandas as pd
mydata = pd.read_csv('mycsv.csv', sep=';', header=0)

x = mydata['X']
y = mydata['Y']

#Definisikan titik data yang akan di-interpolasi
#Gunakan finer and regular mesh
n = 10
xfine = np.linspace(x[0], x[len(mydata) - 1], n*100)

# interpolasi dengan fungsi linear (p=1)
y1 = scipy.interpolate.interp1d(x, y, kind = 'linear')
y1(xfine)
#interpolasi dengan fungsi kuadratik (p=2)
y2 = scipy.interpolate.interp1d(x, y, kind = 'quadratic')
y2(xfine)
#interpolasi dengan fungsi kubik (p=3)
y3 = scipy.interpolate.interp1d(x, y, kind = 'cubic')
y3(xfine)
# Plot data dan hasil interpolasi (y1, y2, dan y3)
plt.plot(x,y,'o', label = 'data point')
plt.plot(xfine,y1(xfine),label = 'linear')
plt.plot(xfine,y2(xfine),label = 'quadratic')
plt.plot(xfine,y3(xfine),label = 'cubic')
plt.legend()
plt.xlabel('x')
plt.xlabel('y')
plt.show()
```

```python
x = [3, 4, 5, 6, 7, 8]
y = [0, 7, 17, 26, 35, 45]
n = len(x)                                  # jumlah data
sumx = sumx2 = sumxy = sumy = 0             # memberi nilai awal nol pada variabel
for i in range(n):
 sumx += x[i]                               # menjumlahkan elemen list x
 sumx2 += x[i]**2                           # menjumlahkan elemen kuadrat list x
 sumxy += x[i]*y[i]                         # menjumlahkan elemen list x dan dikalikan dengan elemen list y
 sumy += y[i]                               # menjumlahkan elemen list y
xm = sumx / n                               # rata-rata nilai x
ym = sumy / n                               # rata-rata nilai y
a = (ym*sumx2 - xm*sumxy)/(sumx2 - n*xm**2) # mencari nilai a
b = (sumxy - xm*sumy)/(sumx2 - n*xm**2)     # mencari nilai b
print('The straight line equation:')
print('f(x) = (%.3f) + (%.3f)x'%(a,b))      # output dari persamaan garis
```

**Latihan 7**

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import pandas as pd

# Interpolasi Linear
def f(x, a1, a0):
 return a1*x + a0
data=pd.read_csv('mycsv.csv',sep=';',header=0)
x = data['X']
y = data['Y']
init_vals = [1, 1]
popt, pcov = curve_fit(f, x, y, p0=init_vals)
print('Best parameter values: {}'.format(popt))
xfitted = np.linspace(x[0], x[len(data)-1], 50)
plt.plot(x, y, 'o', label = 'data $(x_i,y_i)$')
plt.plot(xfitted, f(xfitted, *popt), '-', label = 'fit $f(x_i)$')
plt.title('Plot Regresi Linear')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

# Plot Residual Linear
yres = np.abs(y-f(x, *popt))
plt.plot(x, yres, 'o', label = 'Plot Residual Linear')
plt.ylim(-1, 1)
plt.title('Plot Residual Linear')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Interpolasi Kuadratik
def f2(x, a2, a1, a0):
 return a2*x**2 + a1*x + a0

x = data['X']
y = data['Y']
init_vals = [1, 1, 1]
popt2, pcov = curve_fit(f2, x, y, p0=init_vals)
print('Best parameter values: {}'.format(popt2))
xfitted = np.linspace(x[0], x[len(data)-1], 50)
plt.plot(x, y, 'o', label = 'data $(x_i,y_i)$')
plt.plot(xfitted, f2(xfitted, *popt2), '-', label = 'fit $f(x_i)$')
plt.title('Plot Regresi Kuadratik')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

# Plot Residual Kuadratik
yres = np.abs(y-f2(x, *popt2))
plt.plot(x, yres, 'o', label = 'KPlot Residual Kuadratik')
plt.ylim(-1, 1)
plt.title('Plot Residual Kuadratik')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Interpolasi Kubik
def f3(x, a3, a2, a1, a0):
 return a3*x**3 + a2*x**2 + a1*x + a0

x = data['X']
y = data['Y']
init_vals = [1, 1, 1, 1]
popt3, pcov = curve_fit(f3, x, y, p0=init_vals)
print('Best parameter values: {}'.format(popt3))
xfitted = np.linspace(x[0], x[len(data)-1], 50)
plt.plot(x, y, 'o', label = 'data $(x_i,y_i)$')
plt.plot(xfitted, f3(xfitted, *popt3), '-', label = 'fit $f(x_i)$')
plt.title('Plot Regresi Kubik')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

# Plot Residual Kubik
yres = np.abs(y-f3(x, *popt3))
plt.plot(x, yres, 'o', label = 'Plot Residual Kubik')
plt.ylim(-1, 1)
plt.title('Plot Residual Kubik')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Dari ketiga hasil plot residual yaitu linear, kuadratik, dan kubik dapat disimpulkan bahwa semakin besar nilai variabelnya maka akan plot residu antar titiknya akan semakin sejajar yaitu nilai semakin mendekati 0, yang artinya terdapat pada plot residual kubik memiliki pola yang lebih sejajar dibandingkan plot residual linear dan kuadratik. Oleh karena itu, jenis regresi yang semakin mencerminkan data tersebut yaitu regresi kubik karena titik-titiknya dekat dengan garis horizontal yaitu y = 0
